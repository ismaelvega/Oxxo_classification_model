```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/Clean_DIM_TIENDA_TEST.csv')
df2 = pd.read_csv('../data/Clean_Ventas.csv')
```

```python
df_merged = pd.merge(df1, df2, on="TIENDA_ID", how="inner")  
df_merged.head()
```

```python
umbrales = {
    "BASE": 480000,
    "HOGAR": 490000,
    "PEAATONAL": 420000,
    "RECESO": 516000
}

# Asegurar que ENTORNO_DES no tenga espacios extras también
df_merged["ENTORNO_DES"] = df_merged["ENTORNO_DES"].str.strip()  # estandariza

# Crear columna TARGET
df_merged["TARGET"] = df_merged.apply(
    lambda row: 1 if row["VENTA_TOTAL"] >= umbrales.get(row["ENTORNO_DES"], float('inf')) else 0,
    axis=1
)
```
